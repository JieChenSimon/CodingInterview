````python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class MyLinkedList:
    """修改第 n 个节点的值"""
    
    def __init__(self):
        self.head = None
        self.size = 0
    
    
    def setAtIndex(self, index: int, val: int) -> bool:
        """
        修改第 index 个节点的值为 val (索引从 0 开始)
        
        参数:
            index: 节点位置
            val: 新的值
        
        返回:
            成功返回 True,失败返回 False
        
        时间复杂度: O(n)
        空间复杂度: O(1)
        """
        # 索引无效
        if index < 0 or index >= self.size:
            return False
        
        # 遍历到第 index 个节点
        current = self.head
        for i in range(index):
            current = current.next
        
        # 修改节点的值
        current.val = val
        return True


# ========== 测试用例 ==========

if __name__ == "__main__":
    # 手动创建链表: 1 -> 2 -> 3 -> 4 -> 5
    ll = MyLinkedList()
    ll.head = ListNode(1)
    ll.head.next = ListNode(2)
    ll.head.next.next = ListNode(3)
    ll.head.next.next.next = ListNode(4)
    ll.head.next.next.next.next = ListNode(5)
    ll.size = 5
    
    print("初始链表: 1 -> 2 -> 3 -> 4 -> 5")
    
    # 测试1: 修改头节点
    # result = ll.setAtIndex(0, 10)
    # 上下等价
    result = ll.setAtIndex(index = 0, val = 10)
    print(f"\n修改索引 0 为 10: {result}")
    print(f"修改后头节点值: {ll.head.val}")
    # 预期: True, 头节点值: 10
    
    # 测试2: 修改中间节点
    result = ll.setAtIndex(2, 30)
    print(f"\n修改索引 2 为 30: {result}")
    print(f"修改后索引 2 的值: {ll.head.next.next.val}")
    # 预期: True, 值: 30
    
    # 测试3: 修改尾节点
    result = ll.setAtIndex(4, 50)
    print(f"\n修改索引 4 为 50: {result}")
    print(f"修改后索引 4 的值: {ll.head.next.next.next.next.val}")
    # 预期: True, 值: 50
    
    # 验证整个链表
    print(f"\n验证完整链表:")
    current = ll.head
    index = 0
    while current:
        print(f"索引 {index}: {current.val}")
        current = current.next
        index += 1
    # 预期: 10 -> 2 -> 30 -> 4 -> 50
    
    # 测试4: 无效索引
    result1 = ll.setAtIndex(-1, 99)
    result2 = ll.setAtIndex(10, 99)
    print(f"\n修改索引 -1: {result1}")   # 预期: False
    print(f"修改索引 10: {result2}")     # 预期: False
````

```
## 图解说明
```
修改前: 1 -> 2 -> 3 -> 4 -> 5
索引:   0    1    2    3    4


执行: setAtIndex(2, 30)

步骤:
1. 遍历到索引 2 的节点 (值为 3)
   current = head
   循环: i=0, current移到节点2
   循环: i=1, current移到节点3
2. 修改: current.val = 30

修改后: 1 -> 2 -> 30 -> 4 -> 5
索引:   0    1    2     3    4


````python
#2025.11.16
# 需要明确一点，改链表某个节点的值，不管是头节点，还是非头节点，核心的算法逻辑都是一样的。
#改和删除某个链表节点是不一样的
    # 具体的改的方法要先判断传入值有效性
    # 同样的要先将head保存到一个临时指针中
class ListNode:
    def __init__(self, val = 0, next = None):
        self.val = val
        self.next = next


class MyLinkedList:
    # __init__(self) 不包含 head，说明 初始化时不需要外部传 head 进来
    def __init__(self):
        # 就等于给这个对象加了一个“变量”，叫做 head，你在其他方法中随时能访问
        self.head = None
        # ❌ 不能这样: def setAtIndex(self, index: int, val: int, size: int)
        # 因为 size 不是"传入的参数",而是"对象的状态"
        self.size = 0
    def reviseNodeVal(self, index: int, val: int):
        # 先判断传入值合法性
        if index < 0 or index > self.size:
            return -1
        # 在进入遍历之前先申请一个临时指针保存
        curr = self.head

        # linkedListBefore = []
        for i in range(index):
            # tmp = curr.val
            # linkedListBefore.append(tmp)
            curr = curr.next
        # 这里是遍历到id为index的节点，千万不能再循环中改val值，不然每一个节点的值都会被改
        curr.val = val
        return True



if __name__ == "__main__":

    # 实例化一个链表
    myLinkedList = MyLinkedList()

    # 给链表加值
    myLinkedList.head = ListNode(1)
    myLinkedList.head.next = ListNode(2)
    myLinkedList.head.next.next = ListNode(3)
    myLinkedList.size = 3

    print("linkedListBefore", myLinkedList.head.next.val)
    myLinkedList.reviseNodeVal(index = 1, val = 10)
    print("linkedListAfter", myLinkedList.head.next.val)



    # 对将为节点号为index的值改成val

````
