````python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class MyLinkedList:
    """打印链表"""
    
    def __init__(self):
        self.head = None
    
    
    def printList(self) -> None:
        """
        打印链表
        
        时间复杂度: O(n)
        空间复杂度: O(n) - 存储结果字符串
        """
        result = []
        current = self.head
        
        while current:
            result.append(str(current.val))
            current = current.next
        
        print(" -> ".join(result) + " -> None" if result else "None")


# ========== 测试用例 ==========

if __name__ == "__main__":
    ll = MyLinkedList()
    
    print("测试1: 空链表")
    ll.printList()
    # 预期输出: None
    
    print("\n测试2: 单节点链表")
    ll.head = ListNode(1)
    ll.printList()
    # 预期输出: 1 -> None
    
    print("\n测试3: 多节点链表")
    ll.head = ListNode(1)
    ll.head.next = ListNode(2)
    ll.head.next.next = ListNode(3)
    ll.head.next.next.next = ListNode(4)
    ll.head.next.next.next.next = ListNode(5)
    ll.printList()
    # 预期输出: 1 -> 2 -> 3 -> 4 -> 5 -> None
````
