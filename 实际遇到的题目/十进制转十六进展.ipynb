```python
def manual_dec_to_hex(n: int) -> str:
    if n == 0:
        return '0'
    neg = n < 0
    n = abs(n)
    digits = []
    while n > 0:
        r = n % 16
        if r < 10:
            digits.append(chr(ord('0') + r))
        else:
            digits.append(chr(ord('a') + (r - 10)))  # 小写 a-f
        n //= 16
    if neg:
        digits.append('-')
    return ''.join(reversed(digits))

# 测试
print(manual_dec_to_hex(0))        # '0'
print(manual_dec_to_hex(5)) 
print(manual_dec_to_hex(255))      # 'ff'
print(manual_dec_to_hex(4095))     # 'fff'
print(manual_dec_to_hex(-26))      # '-1a'
```
