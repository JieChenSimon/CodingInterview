#将十进制改成十六进制

def manual_dec_to_hex(n: int) -> str:
    if n == 0:
        return '0'
    neg = n < 0
    n = abs(n)
    digits = []
    while n > 0:
        r = n % 16
        if r < 10:
            digits.append(chr(ord('0') + r))
        else:
            digits.append(chr(ord('a') + (r - 10)))  # 小写 a-f
        n //= 16
    if neg:
        digits.append('-')
    return ''.join(reversed(digits))

# 测试
print(manual_dec_to_hex(0))        # '0'
print(manual_dec_to_hex(5)) 
print(manual_dec_to_hex(255))      # 'ff'
print(manual_dec_to_hex(4095))     # 'fff'
print(manual_dec_to_hex(-26))      # '-1a'


````python
def dec_to_hex(dec):
    if dec == 0:
        return "0"
    hex_digits = "0123456789ABCDEF"
    result = ""
    while dec > 0:
        remainder = dec % 16  # 取余得到当前最低位的十六进制数字
        result = hex_digits[remainder] + result  # 拼接到结果前面，逆序拼接
        dec = dec // 16  # 结合整数除法，用来“缩小”问题规模
    return result
    # return "0x" + result
    # “0x” 是十六进制的识别前缀，不是数值的一部分。

print(dec_to_hex(10))
print(dec_to_hex(128))
````

其中最为重要是：

```python
result = hex_digits[remainder] + result
```
一行的组成与含义

hex_digits 是字符串 "0123456789ABCDEF"，它把 0–15 这 16 个数值映射到对应的十六进制字符。

remainder 是 dec % 16 的结果，是一个整数，范围在 0 到 15 之间。

hex_digits[remainder] 用索引从 hex_digits 取出对应的字符（例如 hex_digits[10] 返回 'A'）。

hex_digits[remainder] + result 是把取到的单个字符与当前 result（也是字符串）连接（字符串拼接）。

整个赋值 result = ... 把新的字符串放回 result，覆盖之前的值。

简短结论：这行把当前最低位（用 remainder 得到）的 十六进制字符放在现有结果的最前面，从而保证最终 result 是从高位到低位的正确顺序。

此外，“0x” 是十六进制的识别前缀，不是数值的一部分。
