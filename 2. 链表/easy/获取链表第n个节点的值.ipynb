## 获取链表第 n 个节点的值

```python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class MyLinkedList:
    """获取链表节点值"""
    
    def __init__(self):
        self.head = None
        self.size = 0
    
    
    def get(self, index: int) -> int:
        """
        获取链表中第 index 个节点的值
        如果索引无效,返回 -1
        
        时间复杂度: O(n)
        空间复杂度: O(1)
        """
        # 边界检查
        if index < 0 or index >= self.size:
            return -1
        
        # 遍历到目标节点
        current = self.head
        for i in range(index):
            current = current.next
        
        return current.val


# ========== 测试用例 ==========

if __name__ == "__main__":
    # 手动创建链表: 1 -> 2 -> 3 -> 4 -> 5
    ll = MyLinkedList()
    ll.head = ListNode(1)
    ll.head.next = ListNode(2)
    ll.head.next.next = ListNode(3)
    ll.head.next.next.next = ListNode(4)
    ll.head.next.next.next.next = ListNode(5)
    ll.size = 5
    
    # 测试获取不同索引的值
    print("测试获取链表节点值:")
    print(f"索引 0 的值: {ll.get(0)}")    # 预期: 1
    print(f"索引 2 的值: {ll.get(2)}")    # 预期: 3
    print(f"索引 4 的值: {ll.get(4)}")    # 预期: 5
    print(f"索引 5 的值: {ll.get(5)}")    # 预期: -1 (超出范围)
    print(f"索引 -1 的值: {ll.get(-1)}")  # 预期: -1 (负数)
```
