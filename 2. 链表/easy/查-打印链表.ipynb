1. 实例化一个链表对象-主函数中
2. 向链表中添加节点 - 主函数
3. 遍历链表并打印每个节点的值- 即新建一个类，包含打印函数

````python
from typing import Optional

# 只负责"节点"的事情
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

#负责"整个链表"的事情，即操作
class MyLinkedList:
    """打印链表"""
    
    def __init__(self):
        self.head = None
    
    
    def printList(self) -> None:
        """
        打印链表
        
        时间复杂度: O(n)
        空间复杂度: O(n) - 存储结果字符串
        """
        result = []
        current = self.head
        
        while current:
            result.append(str(current.val))
            current = current.next
        
        print(" -> ".join(result) + " -> None" if result else "None")


# ========== 测试用例 ==========

if __name__ == "__main__":
    ll = MyLinkedList()
    
    print("测试1: 空链表")
    ll.printList()
    # 预期输出: None
    
    print("\n测试2: 单节点链表")
    ll.head = ListNode(1)
    ll.printList()
    # 预期输出: 1 -> None
    
    print("\n测试3: 多节点链表")
    ll.head = ListNode(1)
    ll.head.next = ListNode(2)
    ll.head.next.next = ListNode(3)
    ll.head.next.next.next = ListNode(4)
    ll.head.next.next.next.next = ListNode(5)
    ll.printList()
    # 预期输出: 1 -> 2 -> 3 -> 4 -> 5 -> None
````

````python
#1. 实例化一个链表对象
#2. 向链表中添加节点
#3. 遍历链表并打印每个节点的值
#核心思想：链表中的printList函数的遍历判断，通过一个临时变量current来遍历链表，直到current为None时结束循环。
class ListNode:
    def __init__(self, val = 0, next = None):
        self.val = val
        self.next = next

class MyLinkedList:
    def __init__(self):
        self.head = None
    def printList(self):
        #临时指针来说存链表头
        curr = self.head
        #new一个数组来存值
        results = []
        # while curr is not None:
        while curr:
            tmp = curr.val
            results.append(tmp)
            # 更新curr节点，即本来curr从head开始的，把curr的指针移动到下一个节点
            curr = curr.next
        print("The Linked List is:", results)

    
if __name__ == "__main__":
    myLinkedList = MyLinkedList()
    myLinkedList.head = ListNode(1)
    myLinkedList.head.next = ListNode(2)

    myLinkedList.printList()


## 2025-11-16
# ListNode中的参数声明不熟悉
# MyLinkedList中的初始化不熟悉
````
