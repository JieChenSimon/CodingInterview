```python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class MyLinkedList:
    """删除第 n 个节点"""
    
    def __init__(self):
        self.head = None
        self.size = 0
    
    
    def deleteAtIndex(self, index: int) -> None:
        """
        删除第 index 个节点(索引从 0 开始)
        
        参数:
            index: 要删除节点的位置
        
        时间复杂度: O(n)
        空间复杂度: O(1)
        """
        # 索引无效
        if index < 0 or index >= self.size:
            return
        
        # 删除头节点
        if index == 0:
            self.head = self.head.next
            self.size -= 1
            return
        
        # 删除非头节点
        # 找到 index-1 位置的节点
        current = self.head
        for i in range(index - 1):
            current = current.next
        
        # 跳过要删除的节点
        current.next = current.next.next
        self.size -= 1


# ========== 测试用例 ==========

if __name__ == "__main__":
    # 手动创建链表: 1 -> 2 -> 3 -> 4 -> 5
    ll = MyLinkedList()
    ll.head = ListNode(1)
    ll.head.next = ListNode(2)
    ll.head.next.next = ListNode(3)
    ll.head.next.next.next = ListNode(4)
    ll.head.next.next.next.next = ListNode(5)
    ll.size = 5
    
    print("初始链表: 1 -> 2 -> 3 -> 4 -> 5")
    print(f"初始大小: {ll.size}")
    
    # 测试1: 删除索引 2 的节点 (删除 3)
    ll.deleteAtIndex(2)
    print(f"\n删除索引 2 后,大小: {ll.size}")
    # 预期: 大小: 4
    print(f"验证: 第1个={ll.head.val}, 第2个={ll.head.next.val}, 第3个={ll.head.next.next.val}")
    # 预期: 1, 2, 4 (跳过了3)
    
    # 测试2: 删除头节点 (删除 1)
    ll.deleteAtIndex(0)
    print(f"\n删除索引 0 后,头节点: {ll.head.val}, 大小: {ll.size}")
    # 预期: 头节点: 2, 大小: 3
    
    # 测试3: 删除尾节点 (删除 5)
    ll.deleteAtIndex(2)
    print(f"\n删除索引 2 后,大小: {ll.size}")
    # 预期: 大小: 2
    print(f"验证: 第1个={ll.head.val}, 第2个={ll.head.next.val}")
    # 预期: 2, 4
    
    # 测试4: 无效索引
    ll.deleteAtIndex(-1)  # 负数索引
    ll.deleteAtIndex(10)  # 超出范围
    print(f"\n删除无效索引后,大小: {ll.size}")
    # 预期: 大小: 2 (没有变化)


```
