```python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class MyLinkedList:
    """在链表头部插入节点"""
    
    def __init__(self):
        self.head = None
        self.size = 0
    
    
    def addAtHead(self, val: int) -> None:
        """
        在链表头部插入节点
        
        时间复杂度: O(1)
        空间复杂度: O(1)
        """
        new_node = ListNode(val)
        new_node.next = self.head
        self.head = new_node
        self.size += 1


# ========== 测试用例 ==========

if __name__ == "__main__":
    ll = MyLinkedList()
    
    print("测试头部插入节点:")
    
    # 测试1: 空链表插入
    ll.addAtHead(1)
    print(f"插入 1 后,头节点值: {ll.head.val}, 大小: {ll.size}")
    # 预期: 头节点值: 1, 大小: 1
    
    # 测试2: 继续插入
    ll.addAtHead(2)
    print(f"插入 2 后,头节点值: {ll.head.val}, 大小: {ll.size}")
    # 预期: 头节点值: 2, 大小: 2
    
    ll.addAtHead(3)
    print(f"插入 3 后,头节点值: {ll.head.val}, 大小: {ll.size}")
    # 预期: 头节点值: 3, 大小: 3
    
    # 验证链表结构: 3 -> 2 -> 1
    print(f"\n验证链表结构:")
    print(f"第1个节点: {ll.head.val}")              # 3
    print(f"第2个节点: {ll.head.next.val}")         # 2
    print(f"第3个节点: {ll.head.next.next.val}")    # 1
```
