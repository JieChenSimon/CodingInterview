```python
from typing import List, Optional
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None


#后面部分是:类型提示，其中的 Optional[ListNode] = 告诉别人"这个值可能是 None,用之前记得检查!"
def createLinkedList(arr: List[int]) -> Optional[ListNode]:
    if arr is None or len(arr) == 0:
        return None

    head = ListNode(arr[0])
    cur = head
    for i in range(1, len(arr)):
        cur.next = ListNode(arr[i])
        cur = cur.next #把cur指向cur.next

    return head
```
