```python
#
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        # 
        fast = 0  # 
        slow = 0  # 
        size = len(nums)
        while fast < size:  # a = size nums[a] 
            # slow  val  fast  val slow 
            if nums[fast] != val:
                nums[slow] = nums[fast]
                slow += 1
            fast += 1
        return slow
```

```python
#
from typing import List
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        i, l = 0, len(nums)
        while i < l:
            if nums[i] == val: # 
                for j in range(i+1, l): # 
                    nums[j - 1] = nums[j]
                l -= 1
                i -= 1
            i += 1
        return l
```
